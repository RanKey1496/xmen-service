```python
directions = [
    {'id': 1, 'detail': 'right'},
    {'id': 2, 'detail': 'left'},
    {'id': 3, 'detail': 'up'},
    {'id': 4, 'detail': 'down'},
    {'id': 5, 'detail': 'upper left'},
    {'id': 6, 'detail': 'lower left'},
    {'id': 7, 'detail': 'upper right'},
    {'id': 8, 'detail': 'lower right'},
]

example_dna = [
    ['A', 'T', 'G', 'C', 'G', 'A'],
    ['C', 'A', 'G', 'T', 'G', 'C'],
    ['T', 'T', 'A', 'T', 'G', 'T'],
    ['A', 'G', 'A', 'A', 'G', 'G'],
    ['C', 'C', 'C', 'C', 'T', 'A'],
    ['T', 'C', 'A', 'C', 'T', 'C']
]
```

```python
def verify_in_directions(directions, x, y, dna):
    try:
        if directions == 1:
            value = f'{dna[x][y]}{dna[x][y+1]}{dna[x][y+2]}{dna[x][y+3]}'
        elif directions == 2:
            value = f'{dna[x][y]}{dna[x][y-1]}{dna[x][y-2]}{dna[x][y-3]}'
        elif directions == 3:
            value = f'{dna[x][y]}{dna[x-1][y]}{dna[x-2][y]}{dna[x-3][y]}'
        elif directions == 4:
            value = f'{dna[x][y]}{dna[x+1][y]}{dna[x+2][y]}{dna[x+3][y]}'
        elif directions == 5:
            value = f'{dna[x][y]}{dna[x-1][y-1]}{dna[x-2][y-2]}{dna[x-3][y-3]}'
        elif directions == 6:
            value = f'{dna[x][y]}{dna[x+1][y-1]}{dna[x+2][y-2]}{dna[x+3][y-3]}'
        elif directions == 7:
            value = f'{dna[x][y]}{dna[x-1][y+1]}{dna[x-2][y+2]}{dna[x-3][y+3]}'
        elif directions == 8:
            value = f'{dna[x][y]}{dna[x+1][y+1]}{dna[x+2][y+2]}{dna[x+3][y+3]}'
        else:
            raise Exception('Se jodiÃ³')
        print(value)
        return value == 'AAA' or value == 'TTTT' or value == 'GGGG' or value == 'CCCC'
    except:
        return False
```

```python
def is_mutant(dna, directions):
    count = 0
    width = len(example_dna)
    max_index = width * width
    for i in range(max_index):
        row = int(i / width)
        column = int(i % width)
        print(row, column)
        for d in directions:
            if verify_in_directions(d['id'], row, column, dna):
                count=count + 1
                print(d['id'], row, column)
    if count > 1:
        print('Es mutante')
        return True
    print('No es mutante')
    return False

%time _ = is_mutant(example_dna, directions)
```
